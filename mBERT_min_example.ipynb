# mBERT on WikiNEuRal dataset for NER

```python
import os
os.environ["USE_TF"] = "0"
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

# === for debu & deprecation warnings ===
# import inspect
import warnings
warnings.filterwarnings('ignore', message='.*Python version.*google.*')
warnings.filterwarnings('ignore', message='.*IProgress not found.*')
warnings.filterwarnings('ignore', message='.*`resume_download` is deprecated*')
warnings.filterwarnings('ignore', message='.*`grouped_entities` is deprecated*')
```

```python
tokenizer = AutoTokenizer.from_pretrained("Babelscape/wikineural-multilingual-ner")
model = AutoModelForTokenClassification.from_pretrained("Babelscape/wikineural-multilingual-ner")

nlp = pipeline("ner", model=model, tokenizer=tokenizer, grouped_entities=True)

# Try a few living languages
texts = [
    "My name is Wolfgang and I work at Google in Berlin.",         # English
    "Je m'appelle Marie et je travaille chez Airbus à Toulouse.",  # French
    "Mi chiamo Luigi e lavoro alla Ferrari a Maranello.",          # Italian
]

for text in texts:
    print(f"\nInput: {text}")
    for entity in nlp(text):
        print(f"  {entity['word']:30} → {entity['entity_group']:5}  (score: {entity['score']:.2f})")
```

```python
# same sample as Minimal_Pipeline_Demo

texts_latin =  ['Quousque tandem abutere, Catilina patientia nostra?']
for text in texts_latin:
    print(f"\nInput: {text}")
    for entity in nlp(text):
        print(f"  {entity['word']:30} → {entity['entity_group']:5}  (score: {entity['score']:.2f})")
```
