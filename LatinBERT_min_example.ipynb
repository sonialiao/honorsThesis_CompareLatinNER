```python
# === for deprecation warnings ===
# import inspect
import warnings
warnings.filterwarnings('ignore', message='.*Python version.*google.*')
warnings.filterwarnings('ignore', message='.*IProgress not found.*')

# === Imports ===
import sys
sys.path.insert(0, '.')

# import numpy as np
from LatinNERpipeline import LatinNerPipeline, tokenizer
from transformers import AutoModelForTokenClassification
```

```python
model = AutoModelForTokenClassification.from_pretrained('./Ner-Latin-RANLP/code/Herodotos_trained_lat_BERT_hypopt_params')
my_pipeline = LatinNerPipeline(model=model, tokenizer=tokenizer)
```

```python
test =  [['Quousque', 'tandem', 'abutere', ',', 'Catilina', 'patientia', 'nostra', '?']]
predict = my_pipeline(test, split_on_words=True)
labels = predict[0]['labels']
print(labels)
```

```python
# Alternate: tokenize on whitespace
test =  ['Quousque tandem abutere, Catilina patientia nostra?', 'Vale']
predict = my_pipeline(test, split_on_words=False)
labels = predict[0]['labels']

print(len(labels))
print(len(test[0].split()))
print(labels)
```
